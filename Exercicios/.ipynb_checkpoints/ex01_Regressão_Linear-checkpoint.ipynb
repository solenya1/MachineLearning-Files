```python
# Deletando todas as colunas que n√£o Contribuem para o Algoritmo
df1 = df.drop(["id", "date","zipcode","lat", "long"], axis=1)
```

## Criando um Regress√£o Linear Simples

```python
lr = LinearRegression()
```

```python
# Separando os dados em dados de treino e teste.
X_treino = np.array(df1["sqft_living"], dtype=pd.Series).reshape(-1, 1)
y_treino = np.array(df1["price"], dtype=pd.Series).reshape(-1, 1)

X_teste = np.array(df1["sqft_living"], dtype=pd.Series).reshape(-1, 1)
y_teste = np.array(df1["price"], dtype=pd.Series).reshape(-1, 1)

lr.fit(X_treino, y_treino)
```

```python
resultado = lr.score(X_teste, y_teste) # Determinado o R2
```

```python
# Coeficiente Angular da Reta
coef_ang = lr.coef_
# Coeficiente Linear da Reta
coef_lin = lr.intercept_
```

```python
print("Coeficiente de Determina√ß√£o(R2)::",resultado)
print("Coeficiente Angular ::", coef_ang)
print("Coeficiente Linear ::",coef_lin)
```

## Visualizando os Resultado

```python
# Plotando o gr√°fico
plt.scatter(X_teste, y_teste, label="Data", color="lightgreen")
plt.plot(X_teste, lr.predict(X_teste), color="red", label="Predict")
plt.title("Visualiza√ß√£o do Modelo")
plt.show()
```

# Solu√ß√£o Correta do Exerc√≠cio

```python
pd.set_option('display.max_columns', 21) # Comando Para visualizar todas Colunas
# Abrindo os arquivos 
arquivo = pd.read_csv("/home/gab/Curso_ML/Exercicios/Datasets/kc_house_data.csv")
```

```python
arquivo.head()
```

```python
# Excluindo features irrelevantes
arquivo.drop("id", axis = 1, inplace = True)
arquivo.drop("date", axis = 1, inplace = True)
arquivo.drop("zipcode", axis = 1, inplace = True)
arquivo.drop("lat", axis = 1, inplace = True)
arquivo.drop("long", axis = 1, inplace = True)

```

```python
# Visualizando os Arquivos Novamente agora Formatados
arquivo.head()
```

```python
# Definindo v√°riaveis preditoras e vari√°vel target
y = arquivo["price"]
x = arquivo.drop(["price"], axis=1)
```

```python
# Separando os dados emtreino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3, random_state=10)
```

```python
# Criando o Modelo
modelo = LinearRegression()
modelo.fit(x_treino, y_treino)
```

```python
# Calculando o coeficiente R2
resultado = modelo.score(x_teste, y_teste)
print('Coeficiente de determina√ß√£o (R2) ::', resultado)
```
