```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 21)
dados = pd.read_csv("/home/gab/MachineLearning-Files/Exercicios/Datasets/kc_house_data.csv")
```

```python
dados.head()
```

```python
delet_columns = ["id", "date", "zipcode", "lat", "long"]

for columns in delet_columns:
    dados.drop(columns, axis=1, inplace = True)
```

```python
dados.head()
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

```python
# Separando os dados
y = dados["price"]
x = dados.drop('price', axis=1)
```

```python
from  sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.30, random_state=14)
```

```python
modelo.fit(x_treino, y_treino)
resultado = modelo.score(x_teste, y_teste)
# Coeficiente de Determinação R2
print("O resultado do modelo é ::", resultado)
```

```python
from sklearn.linear_model import Ridge
modeloRidge = Ridge(alpha=10.0)
# Quando eu aumento o alpha eu estou aumentando o peso dos meu coeficientes de regularização
# Não Significa que vai gerar um melhor resultado.
modeloRidge.fit(x_treino, y_treino)
resultado_Ridge = modeloRidge.score(x_teste, y_teste)
print("Resultado modelo Ridge ::", resultado_Ridge)
```
