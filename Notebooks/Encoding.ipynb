### Aplication of the Hot Encoding 

This function below transform all the data in "Area Name" for example in separated columns.|

```python
area_encode = pd.get_dummies(df['Area Name'])
```

```python
area_encode.head()
```

Now it's time to concat the `area_encode` on the df 

```python
concatenado = pd.concat([df, area_encode], axis=1)
concatenado.drop('Area Name', axis=1, inplace=True)
```
