## Alterando o tipo do dado.

Para mudar o tipo do dado basta alterar a columa selecionando chamando a função `astype()` e alterando dentro dela para o tipo desejado, sendo ele qualquer tipo suportado pelo python como `float64`, `int64`,`object`ou `str`.

Como pode ser observado no exemplo a baixo onde eu alterei a coluna **CouncilDistrictCode** de `int`para `object`.

```python
# Alterando o tipo do dado.
# exemplo de utilização:
dados["CouncilDistrictCode"] = dados["CouncilDistrictCode"].astype(object)
```

## Pré-Processamento: Dados missing

```python
# Para descobrir quais dados são faltantes
enulo = dados.isnull()
```

```python
# Dados que estão faltando por coluna 
faltante = dados.isnull().sum()
print(faltante)
```

```python
# Porcentagem de dados que estão faltando por coluna 
faltantes_percentual = (dados.isnull().sum() / len(dados["SecondLargestPropertyUseTypeGFA"]))*100
```

```python
print(faltantes_percentual)
```

```python
# Apagando Colunas com Dados Vazios ou Irrelevante

Colunas_inuteis  = ["City Council Districts", "YearsENERGYSTARCertified",
                   "ThirdLargestPropertyUseType", "ThirdLargestPropertyUseTypeGFA",
                   "Comment", "Outlier", "2010 Census Tracts"]

for Colunas in Colunas_inuteis:
    dados.drop(Colunas, inplace=True, axis=1)
    
```

```python
dados.shape # Verificando a quantidade de colunas atualizadas
```

```python
dados.head()
```

```python
# Substituindo os dados faltantes por álgum valor
# em Eletricity(kwh) Eu substitui pela médias dos valores dentro da coluna
dados["Electricity(kWh)"].fillna(dados["Electricity(kWh)"].mean(), inplace=True)
```

```python
# ENERGYSTARScore todos os dados que estiverem faltando será substituido pela mediana
dados["ENERGYSTARScore"].fillna(dados["ENERGYSTARScore"].median(), inplace=True)
```

```python
# Verificando o tamanho dos dados
len(dados["ENERGYSTARScore"])
```

```python
# Verificando se os dados faltantes foram preenchidos
verificando_ = dados["ENERGYSTARScore"].isnull().sum()
verificando_
```
