## Dados faltantes

Nesse caso, para esse data set os dados que estão abaixo ou igual a 20% iremos substituir pela médias e o outros dados acimas de 20% iremos excluir. Essa não é a regra geral mas se aplica a esse conjunto de dados pois a intenção e observar o funcionamento da Regressão Logistica. 


Para saber exatamente como lidar com dados faltantes indico um pesquisa sobre "*How to deal with missing data*" existe um excelente artigo da [Medium](https://medium.com/@danberdov/dealing-with-missing-data-8b71cd819501) sobre isso. 

```python
data_to_mean = ["q8.7", "q8.12"]
data_to_delet = ["q8.2", "q8.8", "q8.9","q8.10","q8.17", "q8.18","q8.20", "Respondent.ID", "Product", "q1_1.personal.opinion.of.this.Deodorant"]

#Deletenado dados faltantes muito altos.
for colunas in data_to_delet:
    dados.drop(colunas, axis=1, inplace=True)
    
#Substituido dados faltantes pela mediana.
for columns in data_to_mean:
    dados[columns].fillna(dados[columns].median(), inplace=True)
```

```python
# Verificando se todos dados faltantes foram removidos.
faltantes = dados.isnull().sum()
faltantes_percentual = (dados.isnull().sum() / len(dados['Product.ID'])) * 100
print(faltantes_percentual)
```

Os dados estão ok !!!

```python
#  Separando as variáveis entre preditoras e variável target
y = dados['Instant.Liking']
x = dados.drop('Instant.Liking', axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold


# Separando os dados em fold:
stratifiedkfold = StratifiedKFold(n_splits=5)

# Criando o modelo:
modelo = LogisticRegression(max_iter=5000)
resultado = cross_val_score(modelo, x, y, cv = stratifiedkfold)

# Imprimindo a acurácia:
print("Resultado Logistic Regression ::", resultado.mean())
```
