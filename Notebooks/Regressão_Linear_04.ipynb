## Feature Selection : correlação  Correlação 

### Interdependência entre duas ou mais variáveis como:

    - Massa corporal e altura;
    - Consumo de combústivel e distância percorrida;
    - Tempo de estudo e desempenho na avaliação;
    - etc...
    
Dentro os métodos mais comuns temos os índices de Pearson, Spearman e Kendall.

### Coeficiente de Correlação de Pearson

Também chamado de "*Coeficiente de correlação produto-momento*" mede o grau da correlação(e a direção dessa correlação -- se positiva ou negativa) entre duas variáveis. Os valores assumem representação que variam entre -1 e 1.

### Coeficiente de Correlação de Spearman

O coeficiente de correlação de Spearman é indicado para o cálculo da correlação entre variáveis aleatórias x e y relacionadas monotonicamente entre si, mas não necessariamente de maneira linear. Se a relação é linear o método de Pearson é o mais indicado.

![image](https://miro.medium.com/max/542/1*UItO0nCkSMNRZW7SZHMKaA.png)

Se as variáveis aleatórias seguitem, á perfeição, a função y=log(x), o coeficiente de correlação de Pearson não será igual a 1. Já o coeficiente de Spearman será 1, pois reflete a relação que não é linear.

### Coeficiente de Correlaçção tau de Kendall

É uma medida de correlação de postos, ou seja, verifica a semelhança entre as ordens dos dados quando classificadospor cada uma das quantidades.

Recebe este nome em homenagem ao estatístico britânico Maurice Kendall, que o desenvolveu em 1938.

A correlação de Kendall leva em consideração apenas a concordância direcional dos chamados pares concordantes e discordantes.

## Calculando Correlação com Pandas

```python
import pandas as pd
import numpy as np

pd.set_option('display.width', 90)
# Importando os dados
dados = pd.read_csv("/home/gab/MachineLearning-Files/Exercicios/Datasets/diabetes.csv")
dados.head()
```

```python
# Buscando a correlação dos dados pelo método de Pearson
corr_matrix = dados.corr(method = 'pearson')
corr_matrix
```

```python
# Visualizando as Correlações com HeatMap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
sns.heatmap(dados.corr())
plt.show()
```
