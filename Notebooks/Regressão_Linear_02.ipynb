## Treinando e testando o modelo

A função `train_test_split()` vai pegar os dados criados para x e y e vai dividir eles entre dados de treino e dados de test.

O `test_size=30` indica que eu estou separando 30% dos meus dados para teste e o outros 70% para treino.

O `modelo.score(x_test, y_teste)` Vai calcular o coeficiente de determinação R2 em relação aos dados de teste.

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.30)
```

```python
modelo.fit(x_treino, y_treino) # Criando a reta 
```

```python
resultado = modelo.score(x_teste, y_teste) # Determinando o Coeficiente R2
```

```python
print(resultado) # Resultado do Coeficiente de Determinação R2 
```
